```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

## Line plot

```python
x = np.arange(0,10)
y = np.arange(11,21)

```

```python
## plt plot

plt.plot(x,y, 'r', linestyle = 'dashed',linewidth= 2, marker= "*", markersize = 12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d diagram')
plt.show()
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y, "g+--")
```

## Bar plot

```python
x = [2,8, 10]
y = [11,16,9]



x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2)
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
x = [2,8, 10]
y = [11,16,9]



x2 = [3,9,11]
y2 = [6,15,7]

plt.barh(x,y)
plt.barh(x2,y2)
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## Hist plot

```python
a = np.array([22,67,35,45,56,12,67,89,90])
plt.hist(a)
plt.title('Hist plot')
plt.show()
```

## Box plot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
data
```

```python
plt.boxplot(data)
plt.show()
```

## Pie plot

```python
sizes = [215, 130, 245,210]
labels = ['Python','c','java','ruby']
explode = [0,0,0.2,0]
plt.pie(sizes, labels= labels, explode = explode, autopct= '%1.2f%%')
plt.show()
```

# Functional method

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'g')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,3)
plt.plot(x,y,'g')

plt.subplot(2,2,2)
plt.plot(x,y,'b')

plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

## Object oriented method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.2,0.9,0.9])
#add_axes(left,bottom,width,height)
axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.2,0.9,0.9])
#add_axes(left,bottom,width,height)
axes1.plot(x,y, 'r', label = 'Red')

axes2 = fig.add_axes([0.15,0.75,0.3,0.3])
axes2.plot(x,y,'b', label = 'Blue')

axes1.legend(loc=4)
axes2.legend()
```
