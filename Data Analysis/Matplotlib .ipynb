```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

## Line plot

```python
x = np.arange(0,10)
y = np.arange(11,21)

```

```python
## plt plot

plt.plot(x,y, 'r', linestyle = 'dashed',linewidth= 2, marker= "*", markersize = 12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d diagram')
plt.show()
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y, "g+--")
```

## Bar plot

```python
x = [2,8, 10]
y = [11,16,9]



x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2)
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
x = [2,8, 10]
y = [11,16,9]



x2 = [3,9,11]
y2 = [6,15,7]

plt.barh(x,y)
plt.barh(x2,y2)
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## Hist plot

```python
a = np.array([22,67,35,45,56,12,67,89,90])
plt.hist(a)
plt.title('Hist plot')
plt.show()
```

## Box plot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
data
```

```python
plt.boxplot(data)
plt.show()
```
